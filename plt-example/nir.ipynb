```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ed = pd.read_csv("employee.csv")
```

```python
print(ed.columns.values)
```

```python
departments = np.unique(ed['Department'])
employees_count = []
for dep in departments:
    employees_count.append(len(ed[(ed['Department'] == dep)]))
```

## Amount of employees in each department

```python
plt.figure(figsize=(10, 5))
xs = [i for i, _ in enumerate(departments)]
plt.ylabel("Employees count")
plt.xlabel("Department")
plt.bar(xs, employees_count)
plt.xticks(xs, departments)
plt.show()
```

## Median monthly income and years at company

```python
years = ed['YearsAtCompany'].unique()
rd_data = ed.loc[ed['Department'] == 'Research & Development']
income = []
for year in years:
    income.append(rd_data.loc[rd_data['YearsAtCompany'] == year]['MonthlyIncome'].median())
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(years, income)
plt.xlabel('Years at company')
plt.ylabel('Monthly income')
plt.show()
```
